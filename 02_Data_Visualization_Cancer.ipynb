# Data Visualization

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Import the dataset

From the previous file "01_cleaning_data" import the .csv file 

```python
df = pd.read_csv("Breast-Cancer-Wisconsin-data.csv")
df.head()
```

```python
print("The shape of the dataframe: ", df.shape)
```

```python
df.isnull().sum()
```

```python
df.drop("Unnamed: 0", axis = 1, inplace = True)
```

```python
df.set_index("Id")
```

```python
df.columns
```

Let's visualize the relation between faetures
* 'Clump Thickness'
* 'Uniformity of Cell Size',
* 'Uniformity of Cell Shape',

```python
df.head()
```

The columns class has two values:
* 2 -- "Benign"
* 4 -- "Malign"

```python
df.columns
```

The relationship between 
* Clump Thickness
* Uniformity of Cell Size
* Uniformity of cell Shape

```python
fig, axes = plt.subplots(1, 3, sharex = True, figsize = (16, 4))
sns.barplot(y = "Clump Thickness", x = "Class", data = df, palette="deep", ax = axes[0])
axes[0].set_xticklabels(["Benign", "Malign"])
sns.barplot(y = "Uniformity of Cell Size", x = "Class", data = df, palette="deep", ax = axes[1])
axes[1].set_xticklabels(["Benign", "Malign"])
sns.barplot(y = "Uniformity of Cell Shape", x = "Class", data = df, palette="deep", ax = axes[2])
axes[2].set_xticklabels(["Benign", "Malign"])
```

```python
fig, axes = plt.subplots(1, 3, sharex = True, figsize = (16, 4))
sns.violinplot(y = "Clump Thickness", x = "Class", data = df, palette="deep", ax = axes[0])
axes[0].set_xticklabels(["Benign", "Malign"])
sns.violinplot(y = "Uniformity of Cell Size", x = "Class", data = df, palette="deep", ax = axes[1])
axes[1].set_xticklabels(["Benign", "Malign"])
sns.violinplot(y = "Uniformity of Cell Shape", x = "Class", data = df, palette="deep", ax = axes[2])
axes[2].set_xticklabels(["Benign", "Malign"])
```

```python
fig, axes = plt.subplots(1, 3, sharex = True, figsize = (16, 4))
sns.scatterplot(x = "Uniformity of Cell Size", y = "Uniformity of Cell Shape", hue = "Class", data = df, palette="deep", ax = axes[0])
sns.scatterplot(x = "Clump Thickness", y = "Uniformity of Cell Shape", hue = "Class", data = df, palette="deep", ax = axes[1])
sns.scatterplot(x = "Clump Thickness", y = "Uniformity of Cell Size", hue = "Class", data = df, palette="deep", ax = axes[2])
```

the relationship between:
* Marginal Adhesion
* Single Epithelial Cell Size
* Bare Nuclei

```python
fig, axes = plt.subplots(1, 3, sharex = True, figsize = (16, 4))
sns.barplot(y = "Marginal Adhesion", x = "Class", data = df, palette="deep", ax = axes[0])
sns.barplot(y = "Single Epithelial Cell Size", x = "Class", data = df, palette="deep", ax = axes[1])
sns.barplot(y = "Bare Nuclei", x = "Class", data = df, palette="deep", ax = axes[2])
axes[2].set_xticklabels(["Benign", "Malign"])
```

```python
fig, axes = plt.subplots(1, 3, sharex = True, figsize = (16, 4))
sns.violinplot(y = "Marginal Adhesion", x = "Class", data = df, palette="deep", ax = axes[0])
sns.violinplot(y = "Single Epithelial Cell Size", x = "Class", data = df, palette="deep", ax = axes[1])
sns.violinplot(y = "Bare Nuclei", x = "Class", data = df, palette="deep", ax = axes[2])
axes[2].set_xticklabels(["Benign", "Malign"])
```

```python
fig, axes = plt.subplots(1, 3, sharex = True, figsize = (16, 4))
sns.scatterplot(x = "Marginal Adhesion", y = "Single Epithelial Cell Size", hue = "Class", data = df, palette="deep", ax = axes[0])
sns.scatterplot(x = "Marginal Adhesion", y = "Bare Nuclei", hue = "Class", data = df, palette="deep", ax = axes[1])
sns.scatterplot(x = "Single Epithelial Cell Size", y = "Bare Nuclei", hue = "Class", data = df, palette="deep", ax = axes[2])
```

The relationship between:
* Bland Chromatin
* Normal Nucleoli
* Mitoses

```python
fig, axes = plt.subplots(1, 3, sharex = True, figsize = (16, 4))
sns.barplot(y = "Bland Chromatin", x = "Class", data = df, palette="deep", ax = axes[0])
axes[0].set_xticklabels(["Benign", "Malign"])
sns.barplot(y = "Nromal Nucleoli", x = "Class", data = df, palette="deep", ax = axes[1])
axes[1].set_xticklabels(["Benign", "Malign"])
sns.barplot(y = "Mitoses", x = "Class", data = df, palette="deep", ax = axes[2])
axes[2].set_xticklabels(["Benign", "Malign"])
```

```python
fig, axes = plt.subplots(1, 3, sharex = True, figsize = (16, 4))
sns.violinplot(y = "Bland Chromatin", x = "Class", data = df, palette="deep", ax = axes[0])
axes[0].set_xticklabels(["Benign", "Malign"])
sns.violinplot(y = "Nromal Nucleoli", x = "Class", data = df, palette="deep", ax = axes[1])
axes[1].set_xticklabels(["Benign", "Malign"])
sns.violinplot(y = "Mitoses", x = "Class", data = df, palette="deep", ax = axes[2])
axes[2].set_xticklabels(["Benign", "Malign"])
```

```python
fig, axes = plt.subplots(1, 3, sharex = True, figsize = (16, 4))
sns.scatterplot(x = "Bland Chromatin", y = "Nromal Nucleoli", hue = "Class", data = df, palette="deep", ax = axes[0])
sns.scatterplot(x = "Bland Chromatin", y = "Mitoses", hue = "Class", data = df, palette="deep", ax = axes[1])
sns.scatterplot(x = "Nromal Nucleoli", y = "Mitoses", hue = "Class", data = df, palette="deep", ax = axes[2])
```
